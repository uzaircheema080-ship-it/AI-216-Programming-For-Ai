**Task 3:**

```python
temperatures = [25, 32, 18, 40, 22, 29, 35]

normal = 0
high = 0

for temp in temperatures:
    if 15 <= temp <= 30:
        normal += 1
    elif temp > 30:
        high += 1

print(f"Normal: {normal}")
print(f"High: {high}")
```

**Task 4:**

```python
incomes = [55000, 70000, 65000, 80000, 45000]
statuses = ['Unemployed', 'Employed', 'Employed', 'Employed', 'Employed']

qualified = 0

for i in range(len(incomes)):
    if incomes[i] >= 60000 and statuses[i] == 'Employed':
        qualified += 1

print(f"Qualified applicants: {qualified}")
```

**Task 5:**

```python
speeds = [75, 90, 60, 85, 95, 70, 100]

violations = 0

for speed in speeds:
    if speed > 80:
        violations += 1

percentage = (violations / len(speeds)) * 100

print(f"Violations: {violations}")
print(f"Percentage: {percentage:.1f}%")
```
