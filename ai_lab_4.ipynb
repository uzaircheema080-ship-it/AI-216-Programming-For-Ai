```python

locations = [
    (24.8607, 67.0011),
    (31.5204, 74.3587),
    (33.6844, 73.0479)
]

print("Coordinate pairs:")
for location in locations:
    print(location)


latitudes = []
for location in locations:
    latitudes.append(location[0])
print("\nLatitudes:", latitudes)

print("\nTrying to modify a tuple:")
try:
    locations[0][0] = 25.0000
except TypeError as error:
    print("Error:", error)
    print("Tuples cannot be changed - they are immutable")

count = len(locations)
print("\nTotal coordinate pairs:", count)
```

**Task:5**

```python
sales_data = {
    "electronics": [
        ("Laptop", 120000, 3),
        ("Mouse", 1500, 10)
    ],
    "clothing": [
        ("Shirt", 2500, 5),
        ("Shoes", 5000, 2)
    ]
}
category_revenue = {}

for category in sales_data:
    total = 0
    for item in sales_data[category]:
        product_name, price, quantity = item
        total = total + (price * quantity)
    category_revenue[category] = total

print("Revenue by category:", category_revenue)
unique_products = set()

for category in sales_data:
    for item in sales_data[category]:
        product_name, price, quantity = item
        unique_products.add(product_name)

print("Unique products:", unique_products)
print("\n--- REVENUE REPORT ---")
for category in category_revenue:
    print(f"{category}: Rs. {category_revenue[category]}")

print(f"\nTotal Revenue: Rs. {sum(category_revenue.values())}")
print(f"Number of unique products: {len(unique_products)}")
```

**Task:6**

```python

inventory = {
    "electronics": [
        ("Laptop", 120000, 3),
        ("Mouse", 1500, 10),
        ("Keyboard", 3000, 5)
    ],
    "clothing": [
        ("Shirt", 2500, 5),
        ("Shoes", 5000, 0)  # Sold out
    ],
    "books": [
        ("Python Book", 1500, 4),
        ("Novel", 500, 6)
    ]
}
def add_product(category, name, price, qty):
    if category not in inventory:
        inventory[category] = []
    existing = {p[0] for p in inventory[category]}
    if name in existing:
        print(f"{name} already exists!")
        return

    inventory[category].append((name, price, qty))
    print(f"Added {name} to {category}")
def update_qty(category, name, new_qty):
    for i, product in enumerate(inventory[category]):
        if product[0] == name:
            inventory[category][i] = (name, product[1], new_qty)
            print(f"Updated {name} quantity to {new_qty}")
            return
def remove_sold_out():
    for cat in list(inventory.keys()):
        inventory[cat] = [p for p in inventory[cat] if p[2] > 0]
        if not inventory[cat]:
            del inventory[cat]
def get_unique_products():
    unique = set()
    for products in inventory.values():
        for p in products:
            unique.add(p[0])
    return unique
def generate_report():
    report = {}
    for cat, products in inventory.items():
        total = sum(p[1] * p[2] for p in products)
        report[cat] = total
    return report
print("=" * 50)
print("INVENTORY MANAGEMENT SYSTEM")
print("=" * 50)
add_product("electronics", "Camera", 25000, 2)
add_product("toys", "Car", 800, 3)
update_qty("electronics", "Laptop", 2)
remove_sold_out()
unique = get_unique_products()
print(f"\nUnique Products: {unique}")
report = generate_report()
print(f"\nReport Dictionary: {report}")
print("\nFINAL INVENTORY:")
for cat, products in inventory.items():
    print(f"\n{cat}:")
    for p in products:
        print(f"  {p[0]}: Rs.{p[1]} x {p[2]} = Rs.{p[1]*p[2]}")
```
